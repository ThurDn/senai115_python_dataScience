```python
#1 - importar as bibliotecas
#pandas -> vou transformar a tabela
import pandas as pd
#matplotlib -> para plotar o gráfico
import matplotlib.pyplot as plt
```

```python
#2 -  Ler o arquivo .csv
df = pd.read_csv('./bitcoin.csv')
df.head()
```

```python
#3 - apresentar informações básicas do dataset
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
print(df.shape[0])#verificar quantidade de linhas totais na tabela
df.isnull().sum()#verifica se tem linhas vazias e conta a quantidade
```

```python
#4 - limpar as linhas que possuirem dados vazios
df = df.dropna()
#df_generico = df
df.head()
```

```python
df['Timestamp'].dtypes

df_alterado = df #igualdade, então o inverso é verdadeiro o que o df_alterado mudar df tbm mudará
df_alterado['Timestamp'] = pd.to_datetime(df_alterado['Timestamp'], unit = 's')
df_alterado.head()
```

```python
#6 - plotar o gráfico usando as colunas 'Timestamp' e 'Close'.

eixo_x = df['Timestamp']
eixo_y = df['Close']
plt.plot(eixo_x,eixo_y)
plt.title('Preço do bitcoin ao longo dos anos')
plt.xlabel('Anos')
plt.ylabel('Preço $')
```
